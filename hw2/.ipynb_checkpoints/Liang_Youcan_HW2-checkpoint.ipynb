# Basic Network Analysis
## Question 1

How many nodes and links does each network have? (1pt)


```python
filepath_rand = 'c88-2023_links_random - Sheet1.csv' #REPLACE W FILEPATH
filepath_acq = 'c88-2023_links_acquaintances - Sheet1.csv' #REPLACE W FILEPATH
fn_rand = pd.read_csv(filepath_rand, delimiter=',')
fn_acq = pd.read_csv(filepath_acq, delimiter=',')

# print(fn_rand)
# print(fn_acq)
n_rand=len(fn_rand.index.array)
n_acq=len(fn_acq.index.array)
```

```python
#read nodes from network
G_rand = nx.DiGraph()  
for n in range(n_rand):
    G_rand.add_edge(fn_rand['source'].values[n],fn_rand['target'].values[n])

#read nodes from network
G_acq = nx.DiGraph()  
for n in range(n_acq):
    G_acq.add_edge(fn_acq['source'].values[n],fn_acq['target'].values[n])
print(G_acq)

# Extract the largest connected component
def largest_component_directed(G):
    largest_cc = max(nx.weakly_connected_components(G), key=len)
    G_largest_cc = G_rand.subgraph(largest_cc)
    return G_largest_cc

G_rand = largest_component_directed(G_rand)
G_acq = largest_component_directed(G_acq)
print(G_acq)

# Find the number of nodes
num_nodes_rand = G_rand.number_of_nodes()
num_nodes_acq = G_acq.number_of_nodes()

print("The number of nodes in the random network is:", num_nodes_rand)
print("The number of nodes in the acquaintances network is:", num_nodes_acq)
```

# Question 2

What is the maximum in-degree in each network and the node id/s of the student/s that corresponds to it (2pt)

```python
def max_indegrees(G):
    in_degree = dict(G.in_degree())
    max_in_degree= max(in_degree.values())
    print("The maximum in-degree in this network is: ", max_in_degree)
    nodes_with_max_in_degree = [node for node, in_degree in in_degree.items() if in_degree == max_in_degree]
    print("The corresponding node id is: ",  nodes_with_max_in_degree)

max_indegrees(G_rand)
max_indegrees(G_acq)
```

# Question 4

Generate a Small World graph to model the data selecting its parameters (N,k,p). 

Select N (to match the same number of nodes of the data), k the number of neighbors close to the $<K>$ of the data, and p such that the clustering coefficient of the model resembles the data, to that end, from the expression of part 3. Fill the table with the values of the model you generated (1 pt) 


Check out the documentation for Small World Graph here!
https://networkx.org/documentation/stable/reference/generated/networkx.generators.random_graphs.watts_strogatz_graph.html

```python
# For random link 

# Find the average clustering when p = 0, plug in N and k for random and acquaintence
N = G_rand.number_of_nodes()
k = round(5.8)
C0 =  nx.average_clustering(nx.watts_strogatz_graph(N, k, 0 ))

# Use your equation in qn3 to solve for average clustering when c = p
p=1-pow(0.12065527065527067/C0,1/3)
gs_rand =  nx.watts_strogatz_graph(N, k, p )

# Solving for average degree, average shortest path, number of nodes, number of links
print('Average clustering coefficient = ',nx.average_clustering(gs_rand))
print('Average shortest path length = ',nx.average_shortest_path_length(gs_rand))
print ("Number of nodes:" , gs_rand.number_of_nodes())
print ("Number of edges:" , gs_rand.number_of_edges())
print ("2*Edges/N:" , 2*gs_rand.number_of_edges()/gs_rand.number_of_nodes())
print ("Average degree of  directed graph = ",np.mean(list(dict(gs_rand.degree()).values())))
print("N:",N)
print("k:",k)
print("p:",p)

nx.draw_networkx(gs_rand, node_size=120,font_size=12)
```

```python
# For acquaintance 

# Find the average clustering when p = 0, plug in N and k for random and acquaintence
N = G_acq.number_of_nodes()
k = round(4.717948717948718)
C0 =  nx.average_clustering(nx.watts_strogatz_graph(N, k, 0 ))

# Use your equation in qn3 to solve for average clustering when c = p
p=1-pow(0.1021218596218596/C0,1/3)
gs_acq =  nx.watts_strogatz_graph(N, k, p )

# Solving for average degree, average shortest path, number of nodes, number of links
print('Average clustering coefficient = ',nx.average_clustering(gs_acq))
print('Average shortest path length = ',nx.average_shortest_path_length(gs_acq))
print ("Number of nodes:" , gs_acq.number_of_nodes())
print ("Number of edges:" , gs_acq.number_of_edges())
print ("2*Edges/N:" , 2*gs_acq.number_of_edges()/gs_acq.number_of_nodes())
print ("Average degree of  directed graph = ",np.mean(list(dict(gs_acq.degree()).values())))
print("N:",N)
print("k:",k)
print("p:",p)

nx.draw_networkx(gs_acq, node_size=120,font_size=12)
```

# Question 6

In the same figure compare the histograms of the degrees of the empirical network, and the three models generated in steps 2 - 5 (1pt)


```python
# for random

#G0:
degs0 = list(dict(nx.degree(G_rand)).values())
n0, bins0 = np.histogram(degs0, bins = list(range(min(degs0), max(degs0)+1, 1)), density="True")

#G1:
degs1 = list(dict(nx.degree(gs_rand)).values())
n1, bins1 = np.histogram(degs1, bins = list(range(min(degs1), max(degs1)+1, 1)), density="True")

#G2:
degs2 = list(dict(nx.degree(ger_rand)).values())
n2, bins2 = np.histogram(degs2, bins = list(range(min(degs2), max(degs2)+1, 1)), density="True")

#G3:
degs3 = list(dict(nx.degree(Gba_rand)).values())
n3, bins3 = np.histogram(degs3, bins = list(range(min(degs3), max(degs3)+1, 1)), density="True")

#to plot:
plt.figure(figsize=(10,10)) #use once and set figure size

plt.plot(bins0[:-1],n0,'r-', markersize=10, label="Empirical") 
plt.plot(bins1[:-1],n1,'b-', markersize=10, label="Small World") 
plt.plot(bins2[:-1],n2,'g-', markersize=10, label="Erdos Renyi") 
plt.plot(bins3[:-1],n3,'y-', markersize=10, label="Barabasi Albert") 
plt.legend(loc='upper right', shadow=True)
plt.title('Degree Distributions for random')
plt.xlabel('degree, k')
plt.ylabel('P(k)')
```

```python
# for acquaintance

#G0:
degs0 = list(dict(nx.degree(G_acq)).values())
n0, bins0 = np.histogram(degs0, bins = list(range(min(degs0), max(degs0)+1, 1)), density="True")

#G1:
degs1 = list(dict(nx.degree(gs_acq)).values())
n1, bins1 = np.histogram(degs1, bins = list(range(min(degs1), max(degs1)+1, 1)), density="True")

#G2:
degs2 = list(dict(nx.degree(ger_acq)).values())
n2, bins2 = np.histogram(degs2, bins = list(range(min(degs2), max(degs2)+1, 1)), density="True")

#G3:
degs3 = list(dict(nx.degree(Gba_acq)).values())
n3, bins3 = np.histogram(degs3, bins = list(range(min(degs3), max(degs3)+1, 1)), density="True")

#to plot:
plt.figure(figsize=(10,10)) #use once and set figure size

plt.plot(bins0[:-1],n0,'r-', markersize=10, label="Empirical") 
plt.plot(bins1[:-1],n1,'b-', markersize=10, label="Small World") 
plt.plot(bins2[:-1],n2,'g-', markersize=10, label="Erdos Renyi") 
plt.plot(bins3[:-1],n3,'y-', markersize=10, label="Barabasi Albert") 
plt.legend(loc='upper right', shadow=True)
plt.title('Degree Distributions for acquaintance')
plt.xlabel('degree, k')
plt.ylabel('P(k)')
```
