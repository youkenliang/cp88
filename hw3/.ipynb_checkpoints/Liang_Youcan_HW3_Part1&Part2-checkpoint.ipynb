## Question 3c: Calculate percentages based on your variables of interest

```python
pct_total_pop_car = db['total_pop_car'].sum() / total_pop
pct_total_pop_bus = db['total_pop_bus'].sum() / total_pop
pct_total_pop_male_car = db['male_car'].sum() / total_pop
pct_total_pop_male_bus = db['male_bus'].sum() / total_pop
pct_total_pop_female_car = db['female_car'].sum() / total_pop
pct_total_pop_female_bus = db['female_bus'].sum() / total_pop

print("Percentages of people going to work by private car:", pct_total_pop_car)
print("Percentages of people going to work by public transportation:", pct_total_pop_bus)
print("Percentages of male going to work by private car:", pct_total_pop_male_car)
print("Percentages of male going to work by public transportation:", pct_total_pop_male_bus)
print("Percentages of female going to work by private car:", pct_total_pop_female_car)
print("Percentages of female going to work by public transportation:", pct_total_pop_female_bus)

db["pct_total_pop_car"] = db["total_pop_car"] / \
                  (db["total_pop"] + \
                   (db["total_pop"]==0) * 1
                  )

db["pct_total_pop_bus"] = db["total_pop_bus"] / \
                  (db["total_pop"] + \
                   (db["total_pop"]==0) * 1
                  )

db["pct_male_car"] = db["male_car"] / \
                  (db["total_pop"] + \
                   (db["total_pop"]==0) * 1
                  )

db["pct_male_bus"] = db["male_bus"] / \
                  (db["total_pop"] + \
                   (db["total_pop"]==0) * 1
                  )

db["pct_female_car"] = db["female_car"] / \
                  (db["total_pop"] + \
                   (db["total_pop"]==0) * 1
                  )

db["pct_female_bus"] = db["female_bus"] / \
                  (db["total_pop"] + \
                   (db["total_pop"]==0) * 1
                  )

db.head()
```

```python
import libpysal.weights.set_operations as Wsets
from libpysal.weights import Queen, KNN
import seaborn as sns
import pandas as pd
import geopandas 
import numpy
from sklearn.cluster import KMeans, AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
cluster_variables =  [
    'pct_total_pop_car',        #Percentages of people going to work by private car
    'pct_total_pop_bus',        #"Percentages of people going to work by public transportation
    'pct_male_car',             #"Percentages of male going to work by private car
    'pct_male_bus',             #"Percentages of male going to work by public transportation
    'pct_female_car',           #"Percentages of female going to work by private car
    'pct_female_bus',           #"Percentages of female going to work by public transportation
]
```

```python
# Initialise KMeans instance
kmeans = KMeans(n_clusters=5)

# Set the seed for reproducibility
numpy.random.seed(1234)
# Run K-Means algorithm
k5cls = kmeans.fit(db[cluster_variables])
```

```python
k5cls.labels_
```

```python
# Assign labels into a column
db['k5cls'] = k5cls.labels_
# Setup figure and ax
f, ax = plt.subplots(1, figsize=(9, 9))
# Plot unique values choropleth including a legend and with no boundary lines
db.plot(column='k5cls', categorical=True, legend=True, linewidth=0, ax=ax)
# Remove axis
ax.set_axis_off()
# Display the map
plt.show()
```

```python
# Group data table by cluster label and count observations
k5sizes = db.groupby('k5cls').size()
k5sizes
```

```python
_ = k5sizes.plot.bar()
plt.show()
```

```python
# Dissolve areas by Cluster, aggregate by summing, and keep column for area
areas = db.dissolve(by='k5cls', aggfunc='sum')
areas
```

```python
# Group table by cluster label, keep the variables used 
# for clustering, and obtain their mean
k5means = db.groupby('k5cls')[cluster_variables].mean()
k5means.T.round(2)
```

```python
# Group table by cluster label, keep the variables used 
# for clustering, and obtain their descriptive summary
k5desc = db.groupby('k5cls')[cluster_variables].describe()
# Loop over each cluster and print a table with descriptives
for cluster in k5desc.T:
    print('\n\t---------\n\tCluster %i'%cluster)
    print(k5desc.T[cluster].unstack())
```

```python
# Index db on cluster ID
tidy_db = db.set_index('k5cls')
# Keep only variables used for clustering
tidy_db = tidy_db[cluster_variables]
# Stack column names into a column, obtaining 
# a "long" version of the dataset
tidy_db = tidy_db.stack()
# Take indices into proper columns
tidy_db = tidy_db.reset_index()
# Rename column names
tidy_db = tidy_db.rename(columns={
                        'level_1': 'Attribute', 
                        0: 'Values'})
# Check out result
tidy_db.head(40)
```

Kmeans Clustering (Elbow Method and Silhouette Scores)

```python
import sklearn.cluster as cluster
```

```python
cluster_variables
```

```python
df=db[cluster_variables]
df_Short = df[['pct_total_pop_car','pct_total_pop_bus','pct_male_car','pct_male_bus','pct_female_car','pct_female_bus']]
df_Short
```

```python
K=range(1,12)
wss = []
for k in K:
    kmeans=cluster.KMeans(n_clusters=k,init="k-means++")
    kmeans=kmeans.fit(df_Short)
    wss_iter = kmeans.inertia_
    wss.append(wss_iter)
```

```python
mycenters = pd.DataFrame({'Clusters' : K, 'WSS' : wss})
mycenters
```

```python
# Using elbow method to select the correct number of clusters

_ = plt.figure(figsize = (10,7))
_ = plt.plot(range(1,12), wss, linewidth = 2, color = 'blue', marker='+', markersize = 8)
_ = plt.title('Elbow Method', fontsize = 12)
_ = plt.xlabel('Number of Clusters',fontsize = 10)
_ = plt.ylabel('Inertia',fontsize = 10)

n_clusters = 4

_ = plt.axvline(x = n_clusters, linewidth = 2, color = 'red', linestyle = '--')
_ = plt.show()
```

```python
import sklearn.metrics as metrics
from sklearn.metrics import silhouette_score
from sklearn.metrics import silhouette_samples
from sklearn.datasets import make_blobs
from pylab import rcParams
import numpy as np
import matplotlib as mpl
from matplotlib.ticker import FixedLocator, FixedFormatter
```

```python
# Double-checking the correct number of c;lusters is chosen using silhouette scores

for i in range(2,13):
    labels=cluster.KMeans(n_clusters=i,init="k-means++",random_state=200).fit(df_Short).labels_
    print ("Silhouette score for k(clusters) = "+str(i)+" is "
           +str(metrics.silhouette_score(df_Short,labels,metric="euclidean",sample_size=1000,random_state=200)))
```

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=200).fit(df_Short) for k in range(2,13)]

silhouette_scores = [silhouette_score(df_Short, model.labels_)
                    for model in kmeans_per_k[0:]]
silhouette_scores
```

```python
kmeans_per_k[0:]
```

```python
plt.figure(figsize=(12,10))

for k in (3,4,5,6):
    plt.subplot(2,2,k-2)
    
    y_pred = kmeans_per_k[k-1].labels_
    silhouette_coefficients = silhouette_samples(df_Short, y_pred)
    
    padding = len(df_Short)//30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()
        
        color = mpl.cm.viridis(i/k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                         facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs)//2)
        pos += len(coeffs) + padding
        
    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3,5):
        plt.ylabel('Cluster')
        
    if k in(5,6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel('Silhouette Coefficient')
        
    else:
        plt.tick_params(labelbottom=False)
        
    plt.axvline(x=silhouette_scores[k-2], color='red', linestyle='--')
    print('k=',k,' ',silhouette_scores[k-2])
    plt.title('$k={}$'.format(k), fontsize=16)
    
plt.show()
```
