## Downloading B and C variables

ex. B08203_002E, C18130_010E

https://api.census.gov/data/2015/acs/acs5/variables.html

```python
B_C_vars_to_download = {
    "B17006_007E": "Income",  # No vehicle available
    "C27006_001E": "Medicare coverage",  # Adults 18-64 with a disability
    "B02009_001E": "Black",  # Black or AfAm alone or in combo with one or more other races
    "B02008_001E": "White",  # American Indian or Alaska Native alone or in combo with one or more other races
}
B_C_vars_to_download_list = list(B_C_vars_to_download.keys())
B_C_vars_to_download_list

B_C_var_str = ','.join(B_C_vars_to_download_list)
```

## Downloading S variables

ex. S0101_C01_011EA 

This set of variables are not as reliable as B and C, keep that in mind

https://api.census.gov/data/2015/acs/acs5/subject/variables.html



```python
S_vars_to_download = {
    "S0101_C01_011E": "Estimate!!Total!!Total population!!AGE!!45 to 49 years",  # Total population
}
S_vars_to_download_list = list(S_vars_to_download.keys())
S_vars_to_download_list

S_var_str = ','.join(S_vars_to_download_list)
```

## Generate the URLs

Check the links below when you are done!


```python
# Defining the url
url_B_C = "https://api.census.gov/data/{0}/acs/acs5?get={1}&for=tract:*&in=state:{2}&in=county:{3}&in=tract:*&key={4}".format(
        year, B_C_var_str, state, county, key)

url_S = "https://api.census.gov/data/{0}/acs/acs5/subject?get={1}&for=tract:*&in=state:{2}&in=county:{3}&in=tract:*&key={4}".format(
        year, S_var_str, state, county, key)

print (f'{url_B_C}')
print (f'{url_S}')
```

###Here we are doing two things
1. Converting the json output to a dataframe
2. Converting the strings to float values for calculations

```python
# Retrieve data from the Census and converts it into a dataframe
res = requests.get(url_B_C)
acs_det = pd.DataFrame(res.json())
acs_det.columns = acs_det.iloc[0]
acs_det = acs_det.drop(0).reset_index(drop = True)
```

```python
# Converting Census values from string to numeric type to be able to do calculations
col = acs_det.columns.tolist()
col = [i for i in col if i not in ['state', 'county', 'tract']]
acs_det[col] = acs_det[col].astype(float)
```

Some pitfalls

- No data for that year (very common with S variables)
- Unrealistic results (-99999999999) think about what the result can possibly be
- Mismatch of rows when merging (use inner join)
  - pd.merge(left_df, right_df, on='column_name', how='inner')
- Maps might be weird, sometimes the census regions are not properly plotted (think treasure island for SF county)
- Region of interest for Census API and Cenpy should match
  - ie. If I selected Alameda County for my Census API, I should be searching for Alameda County in my cenpy code
